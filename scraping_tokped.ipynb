### Import Libraries

```python
'''
These libraries are used for scraping data from websites and also to group the collected data into a dataframe.
'''
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

### Data Scraping

```python
'''
Collecting Product "Seblak" Data from Tokopedia
Element: tag = div, class="css-1asz3by"
'''

nama_produk = []
harga_produk = []
penjual = []
kota_penjual = []
sold = []
rating = []

driver = webdriver.Firefox()

for i in range(1,3):
    url = "https://www.tokopedia.com/search?navsource=&page={}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=".format(i)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    seblak_toped = soup.find_all("div", {"class": "css-1asz3by"})
    
    nama_produk += [ nama.get_text() for nama in soup.find_all('div', {"class":"prd_link-product-name css-3um8ox"}) ]
    harga_produk += [ harga.get_text() for harga in soup.find_all('div', {"class":"prd_link-product-price css-h66vau"}) ]

    for j in seblak_toped:
        penjual += [ j.find("span", {"class": "prd_link-shop-name css-1kdc32b flip"}).text if j.find("span", {"class": "prd_link-shop-name css-1kdc32b flip"}) else None ]
        kota_penjual += [ j.find("span", {"class": "prd_link-shop-loc css-1kdc32b flip"}).text if j.find("span", {"class": "prd_link-shop-loc css-1kdc32b flip"}) else None ]
        sold += [ j.find("span", {"class": "prd_label-integrity css-1sgek4h"}).text if j.find("span", {"class": "prd_label-integrity css-1sgek4h"}) else None ]
        rating += [ j.find("span", {"class": "prd_rating-average-text css-t70v7i"}).text if j.find("span", {"class": "prd_rating-average-text css-t70v7i"}) else None ]

df = pd.DataFrame()
df['Nama Produk'] = nama_produk
df['Harga Produk'] = harga_produk
df['Penjual'] = penjual
df['Kota Penjual'] = kota_penjual
df['Produk Terjual'] = sold
df['Rating'] = rating

df
```
